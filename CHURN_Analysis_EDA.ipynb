For features with very high number of missing values- it is better to drop those columns as they give very less insight on analysis.

**DATA CLEANING**

1.Create a copy of base data for manipulation & processing

```python
telco_data=telco_base_data.copy()
```

2.TotalCharges shoud be numeric lets convert it into numeric

```python
# What is errors =' coerce in Python?
# It will ignore all non-numeric values. Output: Code #3: Using errors='coerce'. It will replace all non-numeric values with NaN
#isnull basically used for that how many values are null
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

3.As we can see that there are 11 missing values in Totalcharges column.Let'scheck the records

```python
telco_data.loc[telco_data['TotalCharges'].isnull()==True]
```

**4.Missing value Treatment**
since the percentage  compared to the dataset is very low i.e 0.15,so it is safe to ignore them

```python
#Removing the missing values
telco_data.dropna(how='any',inplace=True)
#If you set inplace = True ,
# the dropna method will modify your DataFrame directly.
#That means that if you set inplace = True , dropna will drop all missing values from your original dataset
#how: how takes string value of two kinds only (‘any’ or ‘all’).
#‘any’ drops the row/column if ANY value is Null and ‘all’ drops only if ALL values are null.
```

5.Divide customers into groups based on the tenure e.g. for tenure <12 months:assign a tenure group if 1-12,for tenure between 1 to 2 Yrs,tenure group of 13-24 and so on

```python
#Get the max value of tenure and according to it divide into groups
telco_data['tenure'].max()
```

```python
#group the tenure in the bins of 12 months
labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
telco_data['tenure_group']=pd.cut(telco_data.tenure,range(1,80,12),right=False,labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

6.Remove columns which are not required

```python
#drop column customerID and tenure
telco_data.drop(columns=['customerID','tenure'],axis=1,inplace=True)
```

### **Bivariate Analysis**

```python
new_df1_target0=telco_data.loc[telco_data['Churn']==0]
new_df1_target1=telco_data.loc[telco_data['Churn']==1]
```

```python
def uniplot(df,col,title,hue =None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright')

    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```
