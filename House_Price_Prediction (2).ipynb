```python
import xgboost as xgb # xgb regressor 
from sklearn.linear_model import LinearRegression,Lasso,Ridge 
# comparison between different models with respect to rmse, mse and r2_score values. 
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
x,y=data.iloc[:,:-1],data.iloc[:,-1]
x.head()
```

```python
y.head()
```

```python
data_dmatrix=xgb.DMatrix(data=x,label=y)
data_dmatrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
xg_reg=xgb.XGBRegressor(objective='reg:linear',colsample_bytree=0.3,learning_rate=0.1,max_depth=5,
                       alpha=10,n_estimators=50)
```

objective= to determine the loss function that is used to linear it means Regression

colsample= percentage of feature used per tree 0.3 = 30%

learning_rate= step size

max_depth= determine to how deeply each tree is allow to grow during boosting

n_estimators= no. of trees you want to build

```python
xg_reg.fit(x_train,y_train)
pred1=xg_reg.predict(x_test)
```

```python
#### mse1=metrics.mean_squared_error(y_test,pred1)
rmse1=np.sqrt(metrics.mean_squared_error(y_test,pred1))
r2_score1=metrics.r2_score(y_test,pred1)
print("Mean squared error : ", mse1)
print("Root mean squared error : ", rmse1)
print("r2_score : ", r2_score1*100)
```

```python
params={'objective':'reg:linear','colsample_bytree':0.3,'learning_rate':0.1,
       'max_depth':5,'alpha':10}
cv=xgb.cv(dtrain=data_dmatrix,params=params,nfold=3,num_boost_round=50,early_stopping_rounds=10,
         metrics='rmse',as_pandas=True,seed=123)
```

```python
cv.head()
```

num_boost_round=similar as n_estimator

early_stopping_rounds= for iteration

n_fold= for cross validation

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred2=lr.predict(x_test)
mse2=metrics.mean_squared_error(y_test,pred2)
rmse2=np.sqrt(metrics.mean_squared_error(y_test,pred2))
r2_score2=metrics.r2_score(y_test,pred2)
print("mean squared error : ",mse2)
print("root mean squared error : ",rmse2)
print("r2_score : ",r2_score2*100)
```

```python
lasso=Lasso(alpha=0.01)
lasso.fit(x_train,y_train)
pred3=lasso.predict(x_test)
mse3=metrics.mean_squared_error(y_test,pred3)
rmse3=np.sqrt(metrics.mean_squared_error(y_test,pred3))
r2_score3=metrics.r2_score(y_test,pred3)
print("mean squared error : ",mse3)
print("root mean squared error : ",rmse3)
print("r2_score : ",r2_score3*100)
```

```python
ridge=Ridge(alpha=0.01)
ridge.fit(x_train,y_train)
pred4=ridge.predict(x_test)
mse4=metrics.mean_squared_error(y_test,pred4)
rmse4=np.sqrt(metrics.mean_squared_error(y_test,pred4))
r2_score4=metrics.r2_score(y_test,pred4)
print("mean squared error : ",mse4)
print("root mean squared error : ",rmse4)
print("r2_score : ",r2_score4*100)
```

```python
error_data1=pd.DataFrame({'Model':['XGBRegressor','LinearRegressor','Lasso','Ridge'],
                          'RMSE':[rmse1,rmse2,rmse3,rmse4],'MSE':[mse1,mse2,mse3,mse4],
                         'r2_score':[r2_score1*100,r2_score2*100,r2_score3*100,r2_score4*100]})
error_data1.head()
```

```python
error_data1.plot(x='Model', y=['RMSE', 'MSE','r2_score'], kind='bar')
plt.title('Comparison between Root-Mean-Squared-Error vs Mean-Squared-Error of different models', fontsize=16)
plt.xlabel('Models', fontsize=14)
plt.ylabel('Errors', fontsize=14)
plt.show()
```
